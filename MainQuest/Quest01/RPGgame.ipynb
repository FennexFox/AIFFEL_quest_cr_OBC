```python
import random

class Character:
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        actual_damage = max(damage - self.defense, 0)
        self.health -= actual_damage
        print(f"{self.name}이 {actual_damage}의 피해를 입었습니다. 남은 체력: {self.health}")

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이 {target.name}에게 {damage}의 데미지를 입힙니다.")
        target.take_damage(damage)

class Player(Character):
    def __init__(self, name, level=1, health=100, attack_power=20, defense=5):
        super().__init__(name, level, health, attack_power, defense)
        self.experience = 0

    def gain_experience(self, amount):
        self.experience += amount
        print(f"{self.name}님이 {amount}의 경험치를 획득했습니다. 총 경험치: {self.experience}")

    def level_up(self):
        if self.experience >= 100:
            self.level += 1
            self.experience -= 100
            print(f"{self.name}님이 레벨업 하셨습니다! 현재 레벨: {self.level}")

def battle(player, monster):
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)

        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        player.gain_experience(monster.level * 20)
        player.level_up()
        print("전투 승리!")
    else:
        print("전투 패배..")

def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    player_name = input("플레이어 이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, level in monster_dict.items():
        monster = Character(monster_name, level, health=level * 50, attack_power=level * 10, defense=level * 2)

        print(f"\n{monster_name} 몬스터와의 전투를 시작합니다!")

        battle(player, monster)

        if not player.is_alive():
            print("게임오버")
            break

if __name__ == "__main__":
    main()
```
